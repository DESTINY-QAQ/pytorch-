```python
def any2dec(target,m):
    n=len(target)-1  #用于处理每一位的指数
    sum=0
    for i in range(len(target)):
        if(target[i])=='A':num=10
        elif target[i]=='B':num=11
        elif target[i]=='C':num=12
        elif target[i]=='D':num=13
        elif target[i]=='E':num=14
        elif target[i]=='F':num=15
        else:num=int(target[i])
        sum+=(m**n)*num
        n-=1
    return sum
ans1=any2dec('1A',16)
ans2=any2dec('11111111',2)
print(f"{ans1},{ans2}")
```

```python
def dec2bin_ex(target):
    i=int(target)
    f=target-i
    a=[]
    while(i!=0):
        a.append(i%2) #append在列表末尾添加一个元素
        i=i//2
    a.reverse()
    b=[]
    n=0
    while(f!=0):
        temp=f*2
        b.append(int(temp))
        f=temp-int(temp)
        n+=1
        if(n>=10):
            break
    return a,b
dec2bin_ex(10.625)
```
