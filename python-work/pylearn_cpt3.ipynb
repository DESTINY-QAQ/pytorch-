```python
import matplotlib.pyplot as plt
import numpy as np

def circle_from_center_and_point(center, point_on_circle):
    """
    给定圆心和圆周上一点，绘制圆。
    center: (x0, y0)
    point_on_circle: (x1, y1)
    """
    cx, cy = center
    px, py = point_on_circle
    # 计算半径
    r = np.hypot(px - cx, py - cy)
    
    # 生成圆周上的点
    theta = np.linspace(0, 2 * np.pi, 400)
    x = cx + r * np.cos(theta)
    y = cy + r * np.sin(theta)
    
    # 绘图
    fig, ax = plt.subplots()
    ax.plot(x, y, label='Circle')
    ax.scatter([cx, px], [cy, py], color='red', zorder=5)
    ax.set_aspect('equal')
    ax.legend()
    plt.show()

circle_from_center_and_point(center=(2, 3), point_on_circle=(5, 7))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def circle_from_square_diagonal(p1, p2):
    """
    给定正方形对角线上两个对顶点，绘制该正方形的外接圆。
    p1: (x1, y1)
    p2: (x2, y2)
    """
    x1, y1 = p1
    x2, y2 = p2
    
    # 圆心为对角线中点
    cx = (x1 + x2) / 2
    cy = (y1 + y2) / 2
    # 半径 = 对角线长度的一半
    r = np.hypot(x2 - x1, y2 - y1) / 2
    
    # 生成圆
    theta = np.linspace(0, 2 * np.pi, 100)
    x = cx + r * np.cos(theta)
    y = cy + r * np.sin(theta)
    
    # 绘图
    fig, ax = plt.subplots()
    ax.plot(x, y, label='Circle')
    # 画出对角线作为参考
    ax.plot([x1, x2], [y1, y2], '--', color='gray', label='Diagonal')
    ax.scatter([x1, x2, cx], [y1, y2, cy], color='red', zorder=5)
    ax.set_aspect('equal')
    ax.legend(loc='upper right')
    plt.show()

# 示例用法
circle_from_square_diagonal(p1=(1, 2), p2=(5, 6))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def circle_from_three_points(p1, p2, p3):
    """
    给定圆周上的三个点，绘制外接圆。
    p1, p2, p3: (x, y)
    """
    x1, y1 = p1
    x2, y2 = p2
    x3, y3 = p3
    
    # 计算两条弦中点连线的垂直平分线交点（圆心）
    # 使用公式推导，避免显式解直线方程
    d = 2 * (x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2))
    if abs(d) < 1e-12:
        raise ValueError("三点共线，无法确定圆")
    
    ux = ((x1**2 + y1**2)*(y2 - y3) + (x2**2 + y2**2)*(y3 - y1) + (x3**2 + y3**2)*(y1 - y2)) / d
    uy = ((x1**2 + y1**2)*(x3 - x2) + (x2**2 + y2**2)*(x1 - x3) + (x3**2 + y3**2)*(x2 - x1)) / d
    cx, cy = ux, uy
    r = np.hypot(x1 - cx, y1 - cy)
    
    # 生成圆
    theta = np.linspace(0, 2 * np.pi, 400)
    x = cx + r * np.cos(theta)
    y = cy + r * np.sin(theta)
    
    # 绘图
    fig, ax = plt.subplots()
    ax.plot(x, y, label='Circle')
    ax.scatter([x1, x2, x3], [y1, y2, y3], color='red', zorder=5, label='Given points')
    ax.scatter(cx, cy, color='green', marker='x', s=100, zorder=5, label='Center')
    ax.set_aspect('equal')
    ax.legend(loc='upper right')
    plt.show()

# 示例用法
circle_from_three_points(p1=(0, 0), p2=(4, 0), p3=(2, 3))
```
